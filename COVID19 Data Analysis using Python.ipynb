# COVID19 data analysis using Python

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print ("modules are imported")
```

# Importing COVID19 dataset

```python
covid19_data=pd.read_csv("D:\Projects\Datasets\corona_dataset\cleaned_data.csv")
covid19_data.head(10)
```

# Checking dataset size using shape function

```python
covid19_data.shape
```

# Removing irrelevant columns for the analysis

```python
df_covid=covid19_data.drop(["Recovered","New deaths","New recovered","New cases","Active"],axis=1) 
df_covid.head(5)
```

# Setting the index to country/Region

```python
df_covid.set_index("Country/Region").head(5)
```

# Remove additional uneccessary columns

```python
df_covid19=df_covid.drop(["Deaths","WHO Region"],axis=1)
df_covid19.head(5)
```

## Changing the Values of the Date column to a column header to show the time series of the data

```python
# Perform a pivot operation with 'Date' as column headers and 'Country' as the index
df_pivoted = df_covid19.pivot(index='Country/Region', columns='Date')


df_pivoted.head(5)
```

```python
df_aggregated=df_pivoted.groupby("Country/Region").sum()
df_aggregated.head(15)
```

```python
df_aggregated.shape
```

```python
df_aggregated.loc["US"].plot()
df_aggregated.loc["China"].plot()
df_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
countries=list(df_aggregated.index)
max_infection_rate=[]
for c in countries: 
    max_infection_rate.append(df_aggregated.loc[c].diff().max())
df_aggregated["max_infection_rate"]=max_infection_rate
```

```python
df_aggregated.head()
```

```python
df_max_data=pd.DataFrame(df_aggregated["max_infection_rate"])
```

```python
df_max_data.head()
```

# Importing economic developmet of countries 

```python
economic_data=pd.read_csv("D:\Projects\Datasets\World_economic_data\Economic_development.csv")
economic_data.head(10)
```

```python
economic_data=economic_data.drop(['Country Code','Series Code','2019','2020','2022'],axis=1)
economic_data.head(10)
```

# Change column values to column header using pivot_table method

```python
economic_data_pivoted = economic_data.pivot_table(index='Country',columns='Series Name',aggfunc='max')
economic_data_pivoted.head(10)
```

```python
economic_data_pivoted = economic_data_pivoted.rename(columns=lambda x: x[1] if isinstance(x, tuple) else x)
economic_data_pivoted.head(10)
```

```python
economic_data_pivoted.shape
```

```python
df_max_data.shape
```

# Joining COVID19 and economic development datasets 

```python
joined_data=df_max_data.join(economic_data_pivoted, how="inner")
joined_data.head(10)
```

```python
print(joined_data.head())
```

```python
#Checking the data types before doing correlation. Only floats and integers are relevant for correlation
print(joined_data.dtypes)
```

```python
# List of object columns to convert to float

joined_data_col = ['GDP', 'GDP per capita', 'Population', 'Population growth','Rural population']

# Replace non-numeric values with NaN
for column in joined_data_col:
    joined_data[column] = joined_data[column].replace('..', float('nan'))

# Convert object columns to float
joined_data[joined_data_col] = joined_data[joined_data_col].astype(float)

```

```python
joined_data.corr()
```

# Data Visualisation

```python
#Plotting GDP vs Infection rate
x=joined_data["GDP"]
y=joined_data["max_infection_rate"]
sns.scatterplot(x=x,y=y)

# Normalize the x-axis values to the range of 0-10
normalized_x = np.interp(x, (min(x), max(x)), (0, 10))

# Normalize the y-axis values to the range of 0-10
normalized_y = np.interp(y, (min(y), max(y)), (0, 10))

sns.scatterplot(x=normalized_x, y=normalized_y)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```
